```python
sample_bootstrap = "https://cdn.jsdelivr.net/npm/bootstrap@5.1.3/dist/css/bootstrap.min.css"
app = Dash(external_stylesheets=[sample_bootstrap])  #[dbc.themes.BOOTSTRAP]
app.layout = html.Div(
    [
            dbc.Row(
                [
                    html.Div("World pop & GDP", 
                             style={"font-size":16,
                                    "color":"white",
                                    "bacground-color":"black",
                                    "text-align":"center"})
                ]
            ),
            dbc.Row(
                [
                    dbc.Col(
                        [
                        html.Label("대륙",
                                    style = {"font-size":10}),
                        dcc.Dropdown(id="dropdown1",  
                                    options = [{"label":i, "value":i} for i in continent],
                                    placeholder = continent[1],
                                    style = {"font-size":10, "height":"5px"}
                                    ),
                               
                          ]
                    ),
                    dbc.Col(
                        [
                         html.Label("국가",
                                       style = {"font-size":10}),
                         html.Div(id="dropdown-result",style={"height":"5px"}),   
                        ]
                         
                    ),
                    dbc.Col(
                        [
                         html.Label("년도",
                                       style={"font-sise":10}),
                         dcc.Dropdown(id="dropdown3",
                                        options = [{"label":i, "value":i} for i in year],
                                        placeholder = year[0],
                                        style = {"font-size":10, "height":"5px"})   
                        ]
                    ),
                ]
            ),
            html.P(),
            html.Hr(),
            dbc.Row(
                [
                    dbc.Col(
                        dcc.Graph(id="bar-graph")
                    ),
                    dbc.Col(
                        dcc.Graph(id="pie-graph")
         
                    ),
                ]
            )
    ]
)
```

```python
@app.callback(
    Output("dropdown-result","children"),
    Input("dropdown1","value")
)

def update_country(continent):
    country = pd.read_sql(f"select distinct country from world where continent = '{continent}'",conn)
    country = country.loc[:,"country"].values
    return html.Div([
        dcc.Dropdown(
            id = "dropdown2",
            options = [{"label":i, "value":i} for i in country],
            style={"font-size":10, "height":"5px"}
        )
    ])

@app.callback(
    Output("bar-graph","figure"),
    Input("dropdown1","value"),
    Input("dropdown2","value"),
    Input("dropdown3","value")
)
def update_bargraph(continent, country, year):
    df = pd.read_sql("select * from world",conn)
    dt = df.loc[(df.continent==continent) & (df.country==country),["year","gdpPercap"]]
    fig = px.bar(x=dt.year.values,y=dt.gdpPercap.values)
    return fig

@app.callback(
    Output("pie-graph","figure"),
    Input("dropdown1","value"),
    Input("dropdown2","value"),
    Input("dropdown3","value")
)
def update_piegraph(continent, country, year):
    df = pd.read_sql("select * from world",conn)
    dt = df.loc[(df.continent==continent) & (df.country==country) ,["year","pop"]]
    fig = px.pie(names=dt.year.values,values=dt.loc[:,"pop"].values)
    return fig
```
